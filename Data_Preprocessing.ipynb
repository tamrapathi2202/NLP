```python
df_no_duplicates = data.drop_duplicates(keep='first')
print(df_no_duplicates)
```

```python
df_no_duplicates['lower_headline'] = df_no_duplicates['Headline'].str.lower()
df_no_duplicates.head()
```

```python
def clean_text(text):
    if isinstance(text, str):
        cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
        return cleaned_text.strip()
    else:
        return ''
```

```python
df_no_duplicates['cleaned_text'] = df_no_duplicates['lower_headline'].apply(clean_text)
df_no_duplicates.head()
```

```python
def remove_stopwords(text):
    if isinstance(text, str):
        words = word_tokenize(text)
        print(words)
        stop_words = set(stopwords.words('english'))
        filtered_words = [word for word in words if word.lower() not in stop_words]
        print(filtered_words)
        cleaned_text = ' '.join(filtered_words)
        print(cleaned_text)
        return cleaned_text.strip()
    else:
        return ''
```

```python
import pandas as pd
from nltk.stem import WordNetLemmatizer

data = {'No_stopwords': ['running fast', 'cats playing outside']}
stop_words_removal = pd.DataFrame(data)

lemmatizer = WordNetLemmatizer()

stop_words_removal['Lemmatized'] = stop_words_removal['No_stopwords'].apply(
    lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()])
)

print(stop_words_removal)
```

```python
stop_words_removal['No_stopwords'] = (stop_words_removal['No_stopwords'].str.replace(r'\s+', ' ', regex=True).str.strip())
```

```python
from collections import Counter
all_words = ' '.join(stop_words_removal['No_stopwords']).split()
word_freq = Counter(all_words)
print(word_freq.most_common(10)) 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=1000)
tfidf_matrix = tfidf.fit_transform(stop_words_removal['No_stopwords'])
print(tfidf_matrix.shape)
```
